### Uso de pipelines

https://huggingface.co/docs/transformers/main_classes/pipelines

Un **pipeline de Hugging Face** es una forma f√°cil de usar un modelo para una tarea espec√≠fica.


image.png

```python
from transformers import pipeline

text_generator = pipeline('text-generation', model='gpt2')
output = text_generator('En un futuro cercano, la inteligencia artificial', max_new_tokens=50)
print(output[0]['generated_text'])
```

```python
qa_pipeline = pipeline('question-answering')
contexto = "EDteam es una compa√±ia de educaci√≥n online"
pregunta = "¬øQu√© desarrolla EDteam?"
respuesta = qa_pipeline(question=pregunta, context=contexto)
print(respuesta)
```

```python
summarizer = pipeline('summarization', model='sshleifer/distilbart-cnn-12-6')
texto_largo = '''EDteam es una plataforma de educaci√≥n online en espa√±ol que ofrece
cursos pr√°cticos de programaci√≥n, dise√±o e inteligencia artificial, con rutas de aprendizaje
estructuradas y acceso mediante planes gratuitos o suscripci√≥n, facilitando el aprendizaje
de tecnolog√≠a de forma clara y accesible'''
resumen = summarizer(texto_largo, max_length=50, min_length=25, do_sample=False)
print(resumen[0]['summary_text'])
```
