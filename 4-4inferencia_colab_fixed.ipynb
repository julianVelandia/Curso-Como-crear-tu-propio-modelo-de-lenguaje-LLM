```python
!pip -q install "huggingface_hub>=0.24.5" llama-cpp-python

from huggingface_hub import snapshot_download
from llama_cpp import Llama
import os, glob

REPO_ID = "JulianVelandia/Llama-3.2-1B-acento-ft-gguf"
HF_TOKEN = ":)"
LOCAL_DIR = "/content/hf_model_best"

snapshot_download(repo_id=REPO_ID, repo_type="model", local_dir=LOCAL_DIR, allow_patterns=["model-f16.gguf"], token=HF_TOKEN)
model_path = glob.glob(f"{LOCAL_DIR}/**/model-f16.gguf", recursive=True)[0]

llm = Llama(model_path=model_path, n_ctx=4096, n_threads=4, chat_format="llama-3")

resp = llm.create_chat_completion(
    messages=[
        {"role":"system","content":"Eres un asistente en español, directo y preciso."},
        {"role":"user","content":"¿Cuál es la capital de Colombia?"}
    ],
    temperature=0.2,
    max_tokens=64
)
print("Modelo:", model_path)
print("Respuesta:", resp["choices"][0]["message"]["content"].strip())
```
